```python
import pandas as pd
import matplotlib as plt
```

```python
states = pd.read_csv("us-states.csv")
states.head()
```

```python
states['date'] = pd.to_datetime(states['date'])
```

```python
washington = states[states.state == "Washington"]
washington['new'] = washington.cases.diff()
washington['prev'] = washington['cases'].shift()
washington['percentDiff'] = (washington.new)/washington.prev
washington.head(50)
```

```python
washington.to_csv(path_or_buf = "washington.csv", index = False)
```

```python
washington = washington[washington.date > pd.to_datetime('02/24/20')]
washington.plot(x="date", y = "percentDiff", figsize=(15,10));
```

```python
def plotState(st, post):
    sta = states[states.state == st]
    sta['new'] = sta.cases.diff()
    sta['percentDiff'] = (sta.new)/sta.prev
    sta = sta[sta.date > pd.to_datetime(post)]
    sta.plot(x="date", y = "percentDiff", figsize=(15,10), ylim = (0,1), title = st + " Percent Difference after "+post);
    
```

```python
plotState("New York", "03/8/20")
```

```python
plotState("Texas", "3/1/20")
```

```python
plotState("California", "3/1/20")
```
